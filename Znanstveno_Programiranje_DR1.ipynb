```python
data_1975 = pd.read_csv("finch_beaks_1975.csv")
data_2012 = pd.read_csv("finch_beaks_2012.csv")
data_1975.head(n=7)
data_1975.memory_usage()
```

```python
data_2012.memory_usage()
```

```python
data_1975.drop(columns=["band"])
#data_2012.drop(columns=["band"])
```

```python

data_1975.insert(0, 'Year', '1975')
data_1975
data_2012.insert(0, 'Year', '2012')
data_2012
```

```python
podaci = [data_1975,data_2012]
Data_merged = pd.concat(podaci)
Data_merged
```

```python
Data_merged.rename(columns = {'Beak length, mm': 'blength1'}, inplace = True)  
Data_merged.rename(columns = {'Beak depth, mm': 'bdepth1'}, inplace = True)

Data_merged
```

```python
Data_merged['blength'] = Data_merged["blength"].fillna(Data_merged.pop('blength1'))
Data_merged
```

```python
Data_merged['bdepth'] = Data_merged["bdepth"].fillna(Data_merged.pop('bdepth1'))
Data_merged
```

```python
Data_merged.memory_usage()
```

```python
Data_merged.groupby("species").count()
```

```python
grouped = Data_merged.groupby(Data_merged.species)

Fortis = grouped.get_group("fortis")
Scandens = grouped.get_group("scandens")

Fortis
Scandens
```
